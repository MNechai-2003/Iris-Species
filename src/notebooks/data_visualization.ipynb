```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Примерные данные
data = pd.DataFrame({
    'Класс': ['A', 'A', 'B', 'B', 'C', 'C'],
    'Категория': ['cat1', 'cat2']*3,
    'Количество': [50, 30, 40, 25, 60, 35]
})

# Построение графика
sns.barplot(x='Класс', y='Количество', hue='Категория', data=data)
plt.title("Распределение объектов по классам и категориям")
plt.show()
```

```python
# Через plotly
import plotly.graph_objects as go

labels = ["A", "B", "C", "cat1", "cat2"]
sources = [0, 0, 1, 1, 2, 2]  # классы
targets = [3, 4, 3, 4, 3, 4]  # категории
values = [50, 30, 40, 25, 60, 35]

fig = go.Figure(data=[go.Sankey(
    node=dict(label=labels),
    link=dict(source=sources, target=targets, value=values)
)])
fig.show()
```
